```python
import os
from collections import defaultdict

# Input: directory path from user
directory = input("Enter the directory path: ")

# Dictionary to hold files by extension
files_by_extension = defaultdict(list)

# Check if directory exists
if os.path.isdir(directory):
    # List all files in the directory
    for file in os.listdir(directory):
        if os.path.isfile(os.path.join(directory, file)):
            # Split filename and extension
            name, ext = os.path.splitext(file)
            ext = ext.lower() if ext else 'No Extension'
            files_by_extension[ext].append(file)

    # Display files categorized by extension
    print("\nFiles categorized by extension:")
    for ext, files in files_by_extension.items():
        print(f"{ext}:")
        for f in files:
            print(f"  {f}")
else:
    print("Invalid directory path. Please check and try again.")
```
