```python
# Input: List from the user
user_input = input("Enter elements of the list separated by spaces: ")
user_list = user_input.split()  # Convert input string into a list of strings

# Input: Item to search
item = input("Enter the item to search for: ")

# Search for the item
if item in user_list:
    position = user_list.index(item)  # Get the index (position)
    print(f"Item '{item}' found at position {position}")
else:
    print("Item not found.")
```
