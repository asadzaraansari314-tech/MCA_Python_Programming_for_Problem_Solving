```python
# Dice probability
def dice_probability(favorable_outcomes, total_faces=6):
    return favorable_outcomes / total_faces

# Coin probability
def coin_probability(favorable_outcomes, total_sides=2):
    return favorable_outcomes / total_sides

# Example usage:

# Dice: probability of rolling a 3
print("Probability of rolling 3 on a dice:", dice_probability(1))

# Dice: probability of rolling an even number
print("Probability of rolling an even number:", dice_probability(3))

# Coin: probability of getting heads
print("Probability of getting heads:", coin_probability(1))

# Coin: probability of getting tails
print("Probability of getting tails:", coin_probability(1))
```
