```python
#(a)
# Number of rows
n = 5

for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end='')  # Print numbers on the same line
    print()  # Move to the next line
```

```python
#(b)
# Number of rows
n = 4

for i in range(n):
    # Print spaces
    print(' ' * (n - i - 1), end='')
    # Print stars
    print('*' * (2 * i + 1))
```

```python
#(c)
# Diamond pattern
n = 3  # Number of rows for the top half (excluding the middle)

# Top half
for i in range(1, n + 1):
    # Print leading spaces
    print(' ' * (n - i) * 3, end='')
    # Print stars with spaces
    print('  '.join(['*'] * (2*i - 1)))

# Bottom half
for i in range(n-1, 0, -1):
    # Print leading spaces
    print(' ' * (n - i) * 3, end='')
    # Print stars with spaces
    print('  '.join(['*'] * (2*i - 1)))

```

```python
#(d)
# Size of the square
rows = 4
cols = 3

for i in range(rows):
    for j in range(cols):
        # Print stars for the border, space inside
        if i == 0 or i == rows - 1 or j == 0 or j == cols - 1:
            print('*', end='')
        else:
            print(' ', end='')
    print()  # Move to next line
```

```python
#(e)
# Diamond/star pattern
rows = 2  # Number of rows for top half

# Top part
for i in range(rows, 0, -1):
    print(' ' * i + '*', end='')  # leading space and first star
    if i != rows:                  # print second star only on bottom/top of diamond
        print(' *', end='')
    print()  # Move to next line

# Bottom part
for i in range(1, rows + 1):
    print(' ' * i + '*', end='')
    if i != rows:
        print(' *', end='')
    print()
```
