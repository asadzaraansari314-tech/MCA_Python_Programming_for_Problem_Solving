```python
# Program to display binary, octal, and hexadecimal representations

num = int(input("Enter an integer: "))

print("\nNumber System Representations")
print("-" * 30)
print(f"Decimal      : {num}")
print(f"Binary       : {bin(num)}")
print(f"Octal        : {oct(num)}")
print(f"Hexadecimal  : {hex(num)}")
```
