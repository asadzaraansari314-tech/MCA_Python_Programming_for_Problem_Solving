```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample employee dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Helen'],
    'Age': [25, 30, 45, 35, 28, 50, 40, 32],
    'Department': ['HR', 'IT', 'IT', 'Finance', 'HR', 'Finance', 'IT', 'HR'],
    'Experience': [2, 5, 20, 10, 3, 25, 15, 6],
    'Salary': [50000, 70000, 120000, 90000, 55000, 130000, 110000, 65000]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the style for seaborn
sns.set(style="whitegrid")

# 1. Histogram for Salary
plt.figure(figsize=(8,5))
sns.histplot(df['Salary'], bins=6, kde=True, color='skyblue')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Number of Employees')
plt.show()

# 2. Box plot for Age
plt.figure(figsize=(8,5))
sns.boxplot(y=df['Age'], color='lightgreen')
plt.title('Age Distribution')
plt.ylabel('Age')
plt.show()

# 3. Bar chart showing department-wise employee count
plt.figure(figsize=(8,5))
sns.countplot(x='Department', data=df, palette='pastel')
plt.title('Employee Count by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.show()

# 4. Scatter plot between Experience and Salary
plt.figure(figsize=(8,5))
sns.scatterplot(x='Experience', y='Salary', data=df, hue='Department', s=100, palette='deep')
plt.title('Experience vs Salary')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.show()

# 5. Correlation heatmap
plt.figure(figsize=(6,4))
corr = df[['Age', 'Experience', 'Salary']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
