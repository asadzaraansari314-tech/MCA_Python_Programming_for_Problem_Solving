```python
# Program to reverse only the vowels in a string

string = input("Enter a string: ")
vowels = "aeiouAEIOU"

# Extract vowels from the string
vowel_list = [ch for ch in string if ch in vowels]

# Reverse the vowels
vowel_list.reverse()

result = []
vowel_index = 0

# Replace vowels with reversed vowels
for ch in string:
    if ch in vowels:
        result.append(vowel_list[vowel_index])
        vowel_index += 1
    else:
        result.append(ch)

print("Result:", "".join(result))
```
