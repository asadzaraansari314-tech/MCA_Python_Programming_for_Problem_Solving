```python
import math

# Take input from user
num = int(input("Enter a number: "))

# Square root
sqrt_val = math.sqrt(num)

# Factorial (only for non-negative integers)
fact_val = math.factorial(num)

# Power calculation
base = int(input("Enter base value: "))
exponent = int(input("Enter exponent value: "))
power_val = math.pow(base, exponent)

# Display results
print("\nResults")
print("--------")
print(f"Square root of {num} = {sqrt_val}")
print(f"Factorial of {num} = {fact_val}")
print(f"{base} raised to the power {exponent} = {power_val}")
```
