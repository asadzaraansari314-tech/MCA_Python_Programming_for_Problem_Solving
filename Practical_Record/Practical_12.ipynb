```python
import random

# Create a list of students
students = ["Zareen", "Zoya", "Rida", "Ziya", "Sana", "Sara"]

# Select CR randomly
cr = random.choice(students)

# Shuffle the student list
random.shuffle(students)

# Display results
print("Selected CR (Class Representative):", cr)
print("\nShuffled Student List:")
print(students)
```
