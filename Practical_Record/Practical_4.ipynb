```python
# Program to find Mean, Median and Mode

numbers = list(map(int, input("Enter numbers separated by space: ").split()))

# Mean
mean = sum(numbers) / len(numbers)

# Median
numbers.sort()
n = len(numbers)
if n % 2 == 0:
    median = (numbers[n//2 - 1] + numbers[n//2]) / 2
else:
    median = numbers[n//2]

# Mode
mode = max(numbers, key=numbers.count)

print("Mean =", mean)
print("Median =", median)
print("Mode =", mode)
```
